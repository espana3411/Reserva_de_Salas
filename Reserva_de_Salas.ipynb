```python
from datetime import datetime
```

```python
calendarioprogramado={}
reservasprogramadas={}
class calendario():
  def __init__(self,calendarioprogramado):
    self.calendarioprogramado=calendarioprogramado
  def comprobardisponibilidad(self,fecha,sala):
    if self.calendarioprogramado=={}:
      return False
    for x in self.calendarioprogramado:
      if x==sala:
        for y in self.calendarioprogramado[x]:
          if y==fecha:
            return True
          else:
            return False
      else:
        return False
  def cancelarreserva(self,sala,fecha):
    try:
     if len(self.calendarioprogramado[sala])>1:
      del self.calendarioprogramado[sala][fecha]
     else: del self.calendarioprogramado[sala]
    except:print("fecha no encontrada")
  def nuevareserva(self,nombre,fecha,sala):
    if sala in self.calendarioprogramado:
      self.calendarioprogramado[sala].update({fecha:nombre})
    else:
     self.calendarioprogramado.update({sala:{fecha:nombre}})
  def modificarfecha(self,fecha1,fecha2,sala):
       cita=self.calendarioprogramado[sala][fecha1]
       del self.calendarioprogramado[sala][fecha1]
       self.calendarioprogramado[sala].update({fecha2:cita})
      
b=calendario(calendarioprogramado)
```
